```python
index = df.index
df = shuffle(df)
df.index = index
```

## Missing Values

```python
# define a funciton to view the total and percentage of missing values 
def view_col_with_nans(df):
    mask_percent = df.isnull().mean().sort_values(ascending=False) 
    mask_total = df.isnull().sum().sort_values(ascending=False)
    total = mask_total[mask_total > 0]
    percent = mask_percent[mask_percent > 0] 
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
view_col_with_nans(df)
```

- The mean_price_level is an important feature, so we can't drop the feature.  
- There is 10% of missing data. Since our dataset is small (n=1521), we can't drop the observations either. 
- I will the "pattern submodel" method in the modeling process to address the missing values problem.

## Feature Engineering

### Create interaction column using Google dataset

Create interaction column between 'mean_price_level' and 'mean_rating', since a neighborhood (zipcode) with a price level of 4 and a mean rating of 5 might be more telling than a neighborhood (zipcode) with a price level of 4 and a mean rating of 1. [(Reference)](https://github.com/Shaddyjr/predicting_affluence_using_yelp/blob/master/code/project_notebook.ipynb)

```python
df['price_level*rating'] = df['mean_price_level'] * df['mean_rating']
```

### Create interaction columns using both Google dataset and Income dataset

Some zipcode may high very high number of total user rating counts. This might indicate the that location is in a more densely populated area. However, we don't know for sure. It would be interesting to see when average user rating per capita.

```python
# Create column user ratings counts devided by the population in the given zipcode
df['user_ratings_total_per_capita'] = df['mean_user_ratings_total'] / df['population']
```

The average user rating in a given zipcode has very limited range (1 to 5). But if we weight the rating with the population density, we might be able to amplify the importance of 'mean_rating'.

```python
# Create column rating weighted with population density in the given zipcode
df['rating*population_density'] = df['mean_rating'] * df['population_density_square_miles']
```

### Check correlational relationships between numerical_cols and target

```python
def scatter_plots(df, numerical_cols, target_col):
    '''
    Take a dataframe, a list of numerical columns, a target column as string
    Return a group of scatter plots
    '''
    # Calculate the number of rows
    num_rows = (len(numerical_cols) // 3) + 1
    # Generate a 3 x n subplots frame
    fix, ax = plt.subplots(num_rows, 3, sharey='row', figsize=(15,20))

    # Reference: https://stackoverflow.com/a/434328
    # Define a function to iterate through a list and divide them into chunks
    def chunker(seq, size):
        return (seq[pos:pos + size] for pos in range(0, len(seq), size))
    
    # Iterate through numerical_cols and generate each subplot
    for y, plot_group in enumerate(chunker((numerical_cols), 3)):
        for x, col in enumerate(plot_group):
            sub_ax = ax[y][x]
            plots = sub_ax.scatter(df[col], df[target_col], s=2)
            plots_titles = sub_ax.set_title(col)
    return (plots, plots_titles)
```

```python
scatter_plots(df, cols_type1, 'log_home_price_to_income_ratios')
```

## Check features's correlation with target

```python
def heatmap_numeric_w_dependent_variable(df, dependent_variable):
    '''
    Generate a heatmap of dependent variable's correlation with y
    '''
    plt.figure(figsize=(8, 30))
    g = sns.heatmap(df.corr()[[dependent_variable]].sort_values(by=dependent_variable), 
                    annot=True, 
                    cmap='coolwarm', 
                    vmin=-1,
                    vmax=1) 
    return g
```

```python
heatmap_numeric_w_dependent_variable(df, 'log_home_price_to_income_ratios')
```
